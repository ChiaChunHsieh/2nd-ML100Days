### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

```python
X.shape
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn import decomposition
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X_pca = pca.transform(X)
X_pca
```

```python
fig = plt.figure(figsize=(4, 3))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=30, azim=154)
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y.astype(np.float))

for name, label in [('0', 0), ('1', 1), ('2', 2)]:
    ax.text3D(X_pca[y == label, 0].mean(),
              X_pca[y == label, 1].mean() + 1.5,
              X_pca[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
    
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
